```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
# 그래프를 노트북 안에 그리기 위해 설정
%matplotlib inline

# 필요한 패키지와 라이브러리 가져오기
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False

# 폰트 지정하기
plt.rcParams['font.family'] = 'gulim'
```

```python
school=pd.read_csv('경기도특수학교현황.csv', encoding='cp949')
school.head()
```

```python
school.info()
```

```python
school.drop(columns=['인가년월일','소재지지번주소','소재지우편번호',
                     '연락처정보','홈페이지URL'], inplace=True)
```

```python
school.head()
```

```python
school.rename(columns={'WGS84위도':'위도'}, inplace=True)
school.rename(columns={'WGS84경도':'경도'}, inplace=True)
school.head()
```

```python
school['개교년월일']=pd.to_datetime(school['개교년월일'])
school.info()
```

```python
school['year']=school['개교년월일'].dt.year
school['month']=school['개교년월일'].dt.month
school.head()
```

```python
school['장애영역명'].unique()
```

```python
# 개교를 많이 한 해
school['year'].value_counts().sort_index().to_frame()
```

### >주로 1~3년에 한개의 학교가 개교 함 ###

```python
# 개교를 많이 한 달
school['month'].value_counts().to_frame()
```

### >3월 개학에 맞추어 개교하는 학교가 많음 ###

```python
# 지역별 특수학교의 장애영역
pd.pivot_table(school, index='시군명', columns='장애영역명',
              values='학교명', aggfunc='count')
```

### >지적장애를 위한 특수학교만 지역별로 한두개씩 있고 다른 장애인을 위한 학교는 턱없이 부족한 것을 알 수 있음.

```python
# 특수학교 위치
map=folium.Map(location=[school.위도.mean(),school.경도.mean()], zoom_start=10,
               tiles='Stamen Terrain') 

for i in school.index:
    folium.Circle(
        radius=500, 
        location=[school.위도[i],school.경도[i]],
        popup=folium.Popup(f'{school.학교명[i]}({school.장애영역명[i]})', max_width=200),
        tooltip=school.소재지도로명주소[i],
        color='black',
        fill=True,
        fill_color='crimson'
    ).add_to(map)
title='<h3 align="center" style="font-size:20px">경기도 특수학교 현황</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```

### >지도로 확인해 보아도 학교가 매우 적음을 알 수 있다. 대부분의 학생들은 차량으로 이동해야 할 거리 ###
