```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("d:/data/images/shapes.jpg")
src = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(src)
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
#      
#     
ret, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
binary = cv2.bitwise_not(binary)
#   
#cv2.findContours( ,  ,  )
#,   
contours, hierarchy = cv2.findContours(binary,cv2.RETR_CCOMP, 
                                       cv2.CHAIN_APPROX_NONE)
for i in range(len(contours)):
    cv2.drawContours(src, [contours[i]], 0, (0, 0, 255), 2)
    cv2.putText(src, str(i), tuple(contours[i][0][0]),
                cv2.FONT_HERSHEY_COMPLEX, 0.8, (0, 255, 0), 1)
    print(i, hierarchy[0][i])
```
