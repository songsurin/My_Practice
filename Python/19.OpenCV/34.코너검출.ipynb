```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("d:/data/images/penguin.png")
src = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(src)
plt.axis("off")
plt.show()
```

```python
dst = src.copy()
gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
#cv2.goodFeaturesToTrack
#(이미지, 최대 코너수, 코너 품질, 최소 거리, 마스크, 블록 크기, 해리스 코너 검출기 유/무, 해리스 코너 계수)
corners = cv2.goodFeaturesToTrack(gray, 20, 0.01, 5,blockSize=3, 
                                  useHarrisDetector=True, k=0.03)
for i in corners:
    cv2.circle(dst, (int(i[0][0]),int(i[0][1])), 3, (0, 0, 255), 2)
```

```python
plt.imshow(dst)
plt.title('image1')
plt.axis("off")
plt.show()
```
