```python
# 
#    0,255  
#  
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("d:/data/images/plane.jpg")
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

```python
#  127  0,255  
#                                     
ret, dst = cv2.threshold(image_rgb, 150, 255, cv2.THRESH_BINARY)
print(ret) #
print(dst) #
```

```python
plt.imshow(dst, cmap="gray")
plt.axis("off")
plt.show()
```
