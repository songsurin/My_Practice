```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("d:/data/images/roo.jpg")
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
image_rgb.shape
```

```python
# 절대크기 dsize(width, height)
dst = cv2.resize(img, dsize=(640, 500))
print(dst.shape)
# 상대크기 fx (width), fy (height)
dst2 = cv2.resize(img, dsize=(0, 0), fx=0.5, fy=0.5)
print(dst2.shape)
```

```python
plt.imshow(image_rgb)
plt.title('original')
plt.axis("off")
plt.show()
image_rgb = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title('dst')
plt.axis("off")
plt.show()
image_rgb = cv2.cvtColor(dst2, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title('dst2')
plt.axis("off")
plt.show()
```
