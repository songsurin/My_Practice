```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("d:/data/images/dog.jpg")
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

```python
import numpy as np
#         grabCut  
# : x, y, width, height
rect = (0,30, 750, 510)
#   
mask1 = np.zeros(image_rgb.shape[:2], np.uint8)
# grabCut    
bgdModel = np.zeros((1, 65), np.float64)
fgdModel = np.zeros((1, 65), np.float64)
# grabCut 
cv2.grabCut(image_rgb, #  
            mask1, # 
            rect, # 
            bgdModel, #    
            fgdModel, #    
            5, #  
            cv2.GC_INIT_WITH_RECT) #   
#   0,   1   
mask2 = np.where((mask1==2) | (mask1==0), 0, 1).astype('uint8')
image_rgb_nobg = image_rgb * mask2[:, :, np.newaxis]
```

```python
plt.figure(figsize=(20,6))
plt.subplot(1, 4, 1)
plt.imshow(image_rgb)
plt.axis("off")
plt.subplot(1, 4, 2)
plt.imshow(mask1)
plt.axis("off")
plt.subplot(1, 4, 3)
plt.imshow(mask2)
plt.axis("off")
plt.subplot(1, 4, 4)
plt.imshow(image_rgb_nobg)
plt.axis("off")
plt.show()
```
