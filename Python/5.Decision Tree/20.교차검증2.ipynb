```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
#플로팅을 위해 2개의 변수만 사용(꽃잎의 길이와 폭)
X = data.data[:, 2:]
y = data.target
feature_names = data.feature_names[2:]
```

```python
#graphbiz를 반환해주는 함수
from sklearn import tree
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import Image, display
import pydotplus
def draw_decision_tree(model):
    dot_tata = tree.export_graphviz(model, out_file=None, feature_names=feature_names,
                                    filled=True, rounded=True, special_characters=True)
    graph = pydotplus.graph_from_dot_data(dot_tata)
    return Image(graph.create_png())
```

```python
#경계도 출력 함수
def plot_decision_regions(X, y, model, title):
    resolution = 0.01
    markers = ('s', '^', 'o')
    colors = ('red', 'blue', 'lightgreen')
    cmap = mpl.colors.ListedColormap(colors)
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    # 정방행렬 생성
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), 
                           np.arange(x2_min, x2_max, resolution))
    Z = model.predict(np.array([xx1.ravel(), xx2.ravel()]).T).reshape(xx1.shape)

    # 등고선, 2차원 넘파일 배열
    plt.contour(xx1, xx2, Z, colors='k')
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8, c=[cmap(idx)],
                    marker=markers[idx], s=80, label=cl)
    plt.xlabel(data.feature_names[2])
    plt.ylabel(data.feature_names[3])
    plt.legend(loc='upper left')
    plt.title(title)
    return Z
```

```python
#학습단계를 바꿔가며 결과 확인
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
for i in range(1, 6):
    plt.figure()
    model = DecisionTreeClassifier(max_depth=i, random_state=0).fit(X, y)
    display(draw_decision_tree(model))
    plot_decision_regions(X, y, model, f'Depth {i}')
    plt.show()
    print(confusion_matrix(y, model.predict(X)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y, model.predict(X)))
```

```python
#교차검증(반복문이용)
from sklearn.model_selection import KFold, cross_val_score
cv = KFold(5, shuffle=True, random_state=0)
i = 1
train_scores = []
test_scores = []
for train_index, test_index in cv.split(X, y):
    print(i, 'fold')
    X_train, X_test = X[train_index, ], X[test_index, ]
    y_train, y_test = y[train_index, ], y[test_index, ]
    model.fit(X_train, y_train)
    score = model.score(X_train, y_train)
    print(score)
    train_scores.append(score)
    score = model.score(X_test, y_test)
    print(score)
    test_scores.append(score)
    i += 1

print('\nAverage')
print(np.mean(train_scores))
print(np.mean(test_scores))

print('\nMax')
print(np.max(train_scores))
print(np.max(test_scores))
```

```python
#교차검증(함수)
model = DecisionTreeClassifier(random_state=0)
cross_val_score(model, X, y, scoring='accuracy', cv=cv).mean()
```
