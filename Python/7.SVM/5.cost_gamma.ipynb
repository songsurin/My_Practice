```python
import matplotlib.pyplot as plt
import mglearn
fig, axes = plt.subplots(3, 3, figsize=(15, 10))
# cost : 0.1, 10, 1000
# gamma : 0.1, 1, 10
for ax, C in zip(axes, [-1, 1, 3]):
    # 같은 번호끼리 결합(10^(-1)=0.1, 10^(1)=10, 10^(3)=1000)
    for a, gamma in zip(ax, [-1, 0, 1]):
        mglearn.plots.plot_svm(log_C=C, log_gamma=gamma, ax=a)
        #ncol=4 4개의 셀을 합쳐서 범례 표시
        axes[0, 0].legend(["class 0", "class 1", "class 0 supportvectors", "class 1 support vectors"], ncol=4, loc=(1, 1.2))

# 감마값 커지면 개별 샘플에 대한 영향력이 커짐
```
