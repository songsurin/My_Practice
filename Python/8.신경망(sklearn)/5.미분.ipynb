```python
#x를 중심으로 전후의 차분을 계산(중심차분 또는 중앙차분)
def numerical_diff(f, x):
    h = 1e-4 # 0.0001
    return (f(x+h) - f(x-h)) / (2*h)
#아래와 같은 2차방정식이 있을 때
def function_1(x):
    return 0.01*x**2 + 0.1*x
```

```python
import matplotlib.pyplot as plt
import numpy as np
X=np.arange(0,20,0.1)
y=function_1(X)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(X,y)
plt.show()
```

```python
#x가 1일 때의 미분값
print(numerical_diff(function_1, 1))
#x가 5일 때의 미분값
print(numerical_diff(function_1, 5))
#x가 10일 때의 미분값
print(numerical_diff(function_1, 10))
```

```python
import matplotlib.pylab as plt
def tangent_line(f, x):
    d = numerical_diff(f, x)
    #print(d)
    y = f(x) - d*x
    return lambda t: d*t + y
    
X = np.arange(0.0, 20.0, 0.1)
y = function_1(X)
plt.xlabel("x")
plt.ylabel("f(x)")
```

```python
#x가 5일 때 y는 0.1999999999990898이다.
#x=5에서의 접선
tf = tangent_line(function_1, 5)
y2 = tf(X)
plt.plot(X, y) #곡선
plt.plot(X, y2) #직선
plt.show()
```

```python
#x=10에서의 접선
tf = tangent_line(function_1, 10)
y2 = tf(X)
plt.plot(X, y) #곡선
plt.plot(X, y2) #직선
plt.show()
```
