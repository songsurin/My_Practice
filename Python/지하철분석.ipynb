## 승차/하차 테이블 분리

### 승차 테이블 만들기

```python
df.columns
```

```python
# 공통 컬럼
df_same=df.iloc[:,:3]
# 승차 컬럼만 가져오기
df_on=df.iloc[:,3::2]
```

```python
df_on.columns=df_on.columns.str.split(" ").str[0] # 시간대만 이름 가져오기
```

```python
# 공통컬럼과 승차컬럼 연결하기
df_on=pd.concat([df_same, df_on], axis=1)
df_on
```

## 하차 테이블 만들기

```python
# 공통 컬럼
df_same=df.iloc[:,:3]
# 하차 컬럼만 가져오기
df_off=df.iloc[:,4::2]
```

```python
df_off.columns=df_off.columns.str.split(" ").str[0]
```

```python
# 공통컬럼과 하차컬럼 연결하기
df_off=pd.concat([df_same, df_off], axis=1)
df_off
```


# 강남역의 최근 시간대별 승하차정보 분석

## 강남역의 최근 승차정보 분석

```python
# 강남역의 최근 승차 데이터 불러오기 
df_Gangnam_on=df_on_202108[df_on_202108['지하철역']=='강남'].iloc[:,3:]
df_Gangnam_on
```

```python
# melt: 행과 열을 바꿔줌
df_Gangnam_on=df_Gangnam_on.melt()
df_Gangnam_on 
```

```python
# 컬럼명 변경
df_Gangnam_on.columns=['시간대','승차건수']
df_Gangnam_on.sort_values('승차건수')
```

# 지하철 시간대별, 역별 이용현황 분석

## 시간대별 승차 현황

### 승차정보 집계 데이터 만들기

```python
# df_in_202108카피하여 사용하기
df_on_202108_agg=df_on_202108.copy()
```

```python
df_on_202108_agg
```

```python
# 인덱스 변경('지하철역')
df_on_202108_agg.index=df_on_202108_agg['지하철역']
```

```python
df_on_202108_agg
```

```python
# 컬럼 삭제('사용월','호선명','지하철역')
df_on_202108_agg.drop(columns=['사용월','호선명','지하철역'], inplace=True)
df_on_202108_agg
```

```python
# 행,열 합계
df_on_202108_agg.loc['sum']=df_on_202108_agg.apply('sum',axis=0)
df_on_202108_agg['sum']=df_on_202108_agg.apply('sum',axis=1)
df_on_202108_agg
```

### 시간대별 승차건수 

```python
s_on=df_on_202108_agg.loc['sum'][:-1].sort_values()
s_on
```

```python
s_on=s_on.sort_index()
```

```python
x=s_on.index
y=s_on.values
plt.bar(x,y)
plt.xticks(rotation=90)
plt.show()
```

### 지하철역별 승차건수

```python
df_on_202108_agg['sum'][:-1].sort_values(ascending=False).to_frame()
```

## 시간대별 하차 현황

### 하차정보 집계 테이블 만들기

```python
# df_out_202108카피하여 사용하기
df_off_202108_agg=df_off_202108.copy()
```

```python
# 인덱스 변경('지하철역')
df_off_202108_agg.index=df_off_202108_agg['지하철역']
```

```python
# 컬럼 삭제('사용월','호선명','지하철역')
df_off_202108_agg.drop(columns=['사용월','호선명','지하철역'], inplace=True)
df_off_202108_agg
```

### 시간대별 하차 건수
