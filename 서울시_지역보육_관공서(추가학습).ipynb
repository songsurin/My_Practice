```python
import pandas as pd
public=pd.read_csv('서울시 지역보육 관공서 정보.csv', encoding='cp949')
public.head(1)
```

```python
public=public[['자치구명','관공서명','X좌표값', 'Y좌표값']]
public.head(1)
```

```python
#컬럼명 변경
public.rename(columns={'X좌표값':'경도'}, inplace=True)
public.rename(columns={'Y좌표값':'위도'}, inplace=True)
public.head(1)
```

```python
public.isnull().sum()
```

```python
public.info()
```

```python
public=public.dropna()
```

```python
public.info()
```

```python
map=folium.Map(location=[public.위도.mean(),public.경도.mean()], zoom_start=12,
               tiles='Stamen Terrain') 

for i in public.index:
    folium.Circle(
        radius=200, #지름
        location=[public.위도[i],public.경도[i]],
        popup=folium.Popup(public.자치구명[i], max_width=200),
        tooltip=public.관공서명[i],
        color='red',
        fill=True,
        fill_color='#3186cc'
    ).add_to(map)
title='<h3 align="center" style="font-size:20px">서울시 지역보육 관공서</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```
